Try to make a network visualization using python.

```python
import networkx as nx
import numpy
from matplotlib import pyplot as plt
```

```python
#read in the data
data = numpy.loadtxt('./data/facebook_combined.txt')
```

```python
#Create a network graph
G = nx.Graph()
edge = []
for line in data:
    edge.append(tuple(line))
G.add_edges_from(edge)
```

```python
%matplotlib inline
plt.figure(figsize=(30,30)) 

nx.draw_networkx(G, node_size = 50, with_labels = False)
plt.show()
```
